**1. Program to create an element-wise comparison.**

```python
import numpy as np
x = np.array([3,5,1,2,3])
y = np.array([2,5,3,2,1])
print("Array A")
print(x)
print("\nArray B")
print(y)

print("\nA>B")
print(np.greater(x,y))

print("\nA>=B")
print(np.greater_equal(x,y))

print("\nA<B")
print(np.less(x,y))

print("\nA<=B")
print(np.less_equal(x,y))
```

**4. Program to compute sum of all elements, sum of each column and sum of
each row.**

```python
import numpy as np

x=np.array([[1,0],[0,1]])

print("Array")
print(x)

print("\n Sum of all elements")
print(np.sum(x))

print("\n Sum of each column")
print(np.sum(x, axis=0))

print("\n Sum of each row")
print(np.sum(x, axis=1))
```

**5. Program to check whether two arrays are equal (element wise) or not.**

```python
import numpy as np

nums1 = np.array([2,2,3,2,1])
nums2 = np.array([2,3,4,3,1])

print("Original arrays:")
print(nums1)
print(nums2)

print("\nTest said two arrays are equal (element wise) or not:?")
print(nums1 == nums2)
print(np.equal(nums1, nums2))
```

**6. Program to multiply two given arrays of same size element-by-element.**

```python
import numpy as np

nums1 = np.array([[2,5,2],[1,5,5]])
nums2 = np.array([[5,3,4],[3,2,5]])

print("Array1:")
print(nums1)
print("Array2:")
print(nums2)

print("\nMultiply said  arrays of same size element-by-element:")
print(np.multiply(nums1,nums2))
```
