**2.DataFrame(2-Dimensional)**

```python
import pandas as pd

data = {
    "Name": ["Alice","Bob"],
    "Age":[20,22]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Transpose

import pandas as pd

df = pd.DataFrame({
    "A":[1,2,3],
    "B":[4,5,6]
})
print(df)
df_t=df.T
print(df_t)
```

```python
#product

import pandas as pd

df = pd.DataFrame({
    "A": [2, 3, 4],
    "B": [5, 6, 7]
})

print(df.prod())
```
